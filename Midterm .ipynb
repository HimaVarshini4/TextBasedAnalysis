## Reading the CSV File and Displaying the DataFrame!

```python
# importing pandas library
import pandas as pd

# Giving the path of the CSV file and stored in a variable name file
file = "C:\\Users\\himav\\Downloads\\Education.csv"

# Reading the CSV file and stored in a variable named called dataframe(df)
df = pd.read_csv(file)

# Printing the first 20 rows of the data from the CSV file
df.head(20)
```

## Function to tokenize

```python
# Tokenizing a Function using a function tokenize()
import regex as re
def tokenize(text):
    return re.findall(r'[\w-]*[a-z|A-Z][\w-]*', text)

x = tokenize('hima Varshini')
print(x)
```

## Remove stop words

```python
# We will use the list of English stop words that are standard in python
import nltk
nltk.download('stopwords')
stopwords = set(nltk.corpus.stopwords.words('english'))
print(stopwords)
```

## Example

```python
# Try it out before running it on the whole dataset
testtext = "Let's defeat SARS-CoV-2 together in 2020! www.cdc.gov"
testtokens = tokenize(testtext)
print(testtokens)
```

## Function to remove stopwords

```python
# Removing the stop words using a function remove_stop()
def remove_stop(tokens):
    return [t for t in tokens if t.lower() not in stopwords]

#Just testing it
cleantokens= remove_stop(testtokens)
print(cleantokens)
```

## Let's make a "pipeline"

```python
# Making a Pipeline
pipeline = [str.lower, tokenize, remove_stop]
def prepare(text, pipeline):
    tokens = text
    for fun in pipeline:
        tokens = fun(tokens)
    return tokens
```

## Run the Pipeline

```python
# Running a pipeline
df['tokens'] = df['Text'].apply(prepare, pipeline=pipeline)
```

## Count the tokens

```python
# Counting the number of tokens using length method
df['num_tokens'] = df['tokens'].map(len)
df.tail()
```

## Use the "Counter"

```python
# Using the Counter method
from collections import Counter

tokens = tokenize("She likes my cats and my cats like my sofa.")
testcount= Counter(tokens)
print(testcount)
```

## Add new data to the current count

```python
# Adding the new data to the current count
more_tokens = tokenize("She likes dogs and cats.")
testcount.update(more_tokens)
print(testcount)
```

## Let's count

```python
# Counting the most common 5 tokens using map() method
counter = Counter()
df['tokens'].map(counter.update)
print(counter.most_common(5))
```

## put the word frequency in it's own dataframe for analysis

```python
# put the word frequency in it’s own dataframe for analysis
freq_df = pd.DataFrame.from_dict(counter,orient='index', columns=['freq'])
freq_df = freq_df.query('freq >= 1')
freq_df.index.name = 'token'
```

## Sort and display!

```python
# Sorting and Displaying 10 values of the frequency dataframe
freq_df = freq_df.sort_values('freq', ascending=False)
freq_df.head(10)
```

## Stats!

```python
# Describing the data of the frequency dataframe
freq_df.describe()
```

## Bar Chart

```python
# plotting top 15 frequency dataframe names in the Y-axis named Token
ax = freq_df.head(15).plot(kind='barh', width=0.95, figsize=(12,8))
ax.invert_yaxis()

# Taking X-axis as 'Frequency', Y-axis as 'Token' and title as 'Top Words'
ax.set(xlabel='Frequency', ylabel='Token', title='Top Words')
```

## Regex - 1 :- Phone Number

```python
# Python code snippet that extracts phone numbers

import re

text = "Hello, my phone number is 123-456-7890."
phone_pattern = r'\b\d{3}-\d{3}-\d{4}\b'

phone_numbers = re.findall(phone_pattern, text)

print("Phone Numbers:", phone_numbers)
```

## Regex - 2 :- URL

```python
# Python code snippet that extracts url's
import re

def find_urls(text):
    url_pattern = r'\bhttps?://\S+\b'
    urls = re.findall(url_pattern, text)
    return urls

# Example usage:
text = "Visit our website at https://example.com for more information."
urls_found = find_urls(text)
print("URLs found:", urls_found)
```

## Regex - 3 :- Email

```python
# Python code snippet that extracts emails
import re

def find_emails(text):
    email_pattern = r'\b[A-Za-z0-9._%+-]+@[A-Za-z0-9.-]+\.[A-Za-z]{2,}\b'
    emails = re.findall(email_pattern, text)
    return emails

# Example usage:
text = "Contact us at email@gmail.com or support@url.org for assistance."
emails_found = find_emails(text)
print("Emails found:", emails_found)
```

```python
# Python code snippet that extracts digits
import re

def find_digits(text):
    digit_pattern = r'\b\d+\b'
    digits = re.findall(digit_pattern, text)
    return digits

# Example usage:
text = "There are 123 apples and 456 oranges."
digits_found = find_digits(text)
print("Digits found:", digits_found)
```

```python
# Python code snippet that extracts alphanumeric sequences
import re

def find_alphanumeric(text):
    alphanumeric_pattern = r'\b[A-Za-z0-9]+\b'
    alphanumeric_sequences = re.findall(alphanumeric_pattern, text)
    return alphanumeric_sequences

# Example usage:
text = "The password is AbCd1234! Don't share it with anyone."
alphanumeric_found = find_alphanumeric(text)
print("Alphanumeric sequences found:", alphanumeric_found)
```
